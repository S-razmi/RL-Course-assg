```python
def render(policy,value):
    import matplotlib.pyplot as plt
    nrows, ncols = 3,4
    values = V.reshape(nrows,ncols)
    plt.figure(figsize=(10,10))
    idx=0
    plt.matshow(values,fignum=1,cmap='YlGnBu_r')
    actions = {0:[0,1],1:[0,-1],2:[1,0],3:[-1,0]}
    for i in range(nrows):
        for j in range(ncols):
            val = round(values[i,j],2)
            arrow= actions[np.argmax(policy[idx])]
            # break_prob = obs[i,j]
            # print(idx,val,np.argmax(policy[idx]),arrow)
            
            plt.text(y=i-0.3,x=j+0.2,s=f'Value {val}',va='center',ha='center',)
            plt.text(y=i+0.3,x=j+0.2,s=idx+1,va='center',ha='center',)
            plt.quiver(j,i,arrow[0],arrow[1],scale=25)
            idx += 1
    plt.show()
```

```python
def reset_values():

    N_STATES = len(states)
    N_ACTIONS = len(actions)
    T = np.zeros((N_STATES, N_ACTIONS, N_STATES)) 
    R = np.zeros((N_STATES, N_ACTIONS, N_STATES))  # rewards
    policy = np.zeros(shape=(12,4))
    policy[:,0] = 1
    policy[[2,4,11]] = 0
    V = np.zeros(N_STATES)

    T[0,0,0] = 1.0
    T[0,1,0] = 1.0
    T[0,2,1] = 1.0
    T[0,3,0] = 1.0

    T[1,0,1] = 1.0
    T[1,1,5] = 1.0
    T[1,2,2] = 1.0
    T[1,3,0] = 1.0

    T[2,0,2] = 1.0
    T[2,1,2] = 1.0
    T[2,2,2] = 1.0
    T[2,3,2] = 1.0

    T[3,0,3] = 1.0
    T[3,1,7] = 1.0
    T[3,2,3] = 1.0
    T[3,3,2] = 1.0

    T[4,0,4] = 1.0
    T[4,1,4] = 1.0
    T[4,2,4] = 1.0
    T[4,3,4] = 1.0

    T[5,0,1] = 1.0
    T[5,1,9] = 1.0
    T[5,3,5] = 1.0
    T[5,2,6] = 1.0

    T[6,0,2] = 1.0
    T[6,1,10] = 1.0
    T[6,2,7] = 1.0
    T[6,3,5] = 1.0

    T[7,0,3] = 1.0
    T[7,1,11] = 1.0
    T[7,3,7] = 1.0
    T[7,2,6] = 1.0

    T[8,0,8] = 1.0
    T[8,1,8] = 1.0
    T[8,2,9] = 1.0
    T[8,3,8] = 1.0

    T[9,0,5] = 1.0
    T[9,1,9] = 1.0
    T[9,2,10] = 1.0
    T[9,3,8] = 1.0

    T[10,0,6] = 1.0
    T[10,1,10] = 1.0
    T[10,3,9] = 1.0
    T[10,2,11] = 1.0

    T[11,0,11] = 1.0
    T[11,1,11] = 1.0
    T[11,2,11] = 1.0
    T[11,3,11] = 1.0

    R[0,0,0] = 0
    R[0,1,0] = 0
    R[0,2,1] = 0
    R[0,3,0] = 0

    R[1,0,1] = 0
    R[1,1,6] = 0
    R[1,2,2] = -10
    R[1,3,0] = 0

    R[2,0,2] = 0
    R[2,1,2] = 0
    R[2,2,2] = 0
    R[2,3,2] = 0

    R[3,0,3] = 0
    R[3,1,7] = 0
    R[3,2,3] = 0
    R[3,3,2] = -10

    R[4,0,4] = 0
    R[4,1,4] = 0
    R[4,3,4] = 0
    R[4,2,4] = 0

    R[5,0,1] = 0
    R[5,1,9] = 0
    R[5,3,5] = 0
    R[5,2,6] = 0

    R[6,0,2] = -10
    R[6,1,10] = 0
    R[6,2,7] = 0
    R[6,3,5] = 0

    R[7,0,3] = 0
    R[7,1,11] = 10
    R[7,2,7] = 0
    R[7,3,6] = 0

    R[8,0,8] = 0
    R[8,1,8] = 0
    R[8,2,9] = 0
    R[8,3,8] = 0

    R[9,0,5] = 0
    R[9,1,9] = 0
    R[9,2,10] = 0
    R[9,3,8] = 0

    R[10,0,6] = 0
    R[10,1,10] = 0
    R[10,3,9] = 0
    R[10,2,11] = 10

    R[11,0,11] = 0
    R[11,1,11] = 0
    R[11,2,11] = 0
    R[11,3,11] = 0
    return R,V,T,policy
```

```python
def get_actions(policy):
    reverse_action = {3:'LEFT',1:'DOWN',2:'RIGHT',0:'UP'}
    print("_______________________________________")
    for i,action in enumerate(np.argmax(policy,axis=1)):
        print(i,reverse_action[action])
```

```python
R,V,T,policy = reset_values()
V[[5,6]] =5
```

```python
render(policy,V)
```

```python
theta=0.1
iteration = 0
gamma=0.9
print(0,V)
print(0,get_actions(policy))
while True:
    iteration+=1
    while True:
        delta=0
        print('_____________________________________________________')
        for s in states:
            if s in [2,4,11]:
                continue
            v_old = V[s]
            G = (gamma * V + R[s])
            V[s] = np.sum((T[s] * G ).T*policy[s])
            delta = max(np.abs(v_old - V[s]),delta)
            print(iteration,s,V)
        if delta<theta:
            break

    policy_stable = True
    print('________________________________________________________')
    for s in states:
        old_policy = np.copy(policy[s])
        if s in [2,4,11]:
            continue
        G = (gamma * V+ R[s])
        Q = np.sum((T[s] * G ).T,axis=0)
        policy[s] = 0
        policy[s][np.argmax(Q)] = 1
        print(iteration,s,get_actions(policy))
        if np.allclose(old_policy,policy[s])==False:
            policy_stable = False
    render(policy,V)
    if policy_stable:
        break
```
