```python
import matplotlib.pyplot as plt
import numpy as np
from strategies import doctor_a,doctor_b,doctor_c
```

```python
A_rewards=[]
B_rewards=[]
C_rewards=[]
for i in range(20):
    a = doctor_a()[-1,2]
    b = doctor_b()[-1,2]
    c = doctor_c()[-1,2]
    A_rewards.append([i,a])
    B_rewards.append([i,b])
    C_rewards.append([i,c])
A_rewards = np.array(A_rewards)
B_rewards = np.array(B_rewards)
C_rewards = np.array(C_rewards)
plt.figure(figsize=(8,8))
plt.title("Last Patient's Reward Boxplot",size=16)
plt.ylabel("Reward")
plt.boxplot(np.stack([A_rewards[:,1],B_rewards[:,1],C_rewards[:,1]],axis=1),labels=["Doctor A","Doctor B","Doctor C"])
plt.grid()
plt.show()
```
