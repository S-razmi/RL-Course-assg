```python
import matplotlib.pyplot as plt
import numpy as np
from strategies import doctor_a,doctor_b,doctor_c
from scipy.stats import t
```

# 5 Repeat

```python
A_rewards=[]

for i in range(5):
    a = doctor_a()
    A_rewards.append(a)

rewards = np.stack([i[:,2] for i in A_rewards],axis=1)
std = np.std(rewards,axis=1)
mean= np.mean(rewards,axis=1)

N=len(A_rewards)
dof = N-1
alpha=0.05

t_crit = np.abs(t.ppf((alpha)/2,dof))
upper_curve = mean+std*t_crit/np.sqrt(N)
lower_curve = mean-std*t_crit/np.sqrt(N)

plt.figure(figsize=(20,8))
plt.title("Doctor A, Cummulative Reward Per Trial",size=20)
plt.ylabel("Reward",size=15)
plt.xlabel("Trial",size=15)
plt.grid()
plt.fill_between(range(1,101),lower_curve,upper_curve,alpha=0.2,label='95% Confidence Interval')
plt.plot(range(1,101),mean,marker='o',linewidth=2,markersize=5,label='Average reward')
plt.legend()
plt.show()
```
