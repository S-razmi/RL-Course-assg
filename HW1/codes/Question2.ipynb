```python
import matplotlib.pyplot as plt
from strategies import doctor_a,doctor_b,doctor_c
```

```python
A_rewards = doctor_a()
B_rewards = doctor_b()
C_rewards = doctor_c()

fig,axs=plt.subplots(nrows=3,ncols=1,figsize=(15,8))

plt.suptitle("Cummulative Reward Per Trial",size=20)

axs[0].set_title('Doctor A:')
axs[0].set_ylabel("Reward")
axs[0].set_xlabel("Trial")
axs[0].grid()
axs[0].plot(A_rewards[:,0],A_rewards[:,2],marker='o',linewidth=1,markersize=3)

axs[1].set_title('Doctor B:')
axs[1].set_ylabel("Reward")
axs[1].set_xlabel("Trial")
axs[1].grid()
axs[1].plot(B_rewards[:,0],B_rewards[:,2],marker='o',linewidth=1,markersize=3,c='g')

axs[2].set_title('Doctor C:')
axs[2].set_ylabel("Reward")
axs[2].set_xlabel("Trial")
axs[2].grid()
axs[2].plot(C_rewards[:,0],C_rewards[:,2],marker='o',linewidth=1,markersize=3,c='r')

fig.tight_layout()
plt.show()
```
